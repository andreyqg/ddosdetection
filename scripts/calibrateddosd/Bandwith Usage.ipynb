```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np

### Statistics per second ###

statsxsec = pd.read_csv("statspersec.txt",delimiter='|',usecols=[3])
#statistic.head()
plt.figure(figsize=(20, 5))
plt.plot(statsxsec,color='blue',lw=.05)
#plt.xlabel('Time (Mins)')
#plt.ylabel('Bandwith Usage (MB/Sec)')
plt.savefig('bandwithpersecond.pdf')
```

```python
### Statistics each 109 seconds, corresponding to 1 observation window ###

### Statistics per observation window ###
statsxow = pd.read_csv("statsperow.txt",delimiter='|',usecols=[3])
#statistic.head()
plt.figure(figsize=(10, 5))
plt.suptitle('Bandwith Utilization per Window', fontsize=20)
plt.plot(statsxow/100000,color='green',lw=1)
plt.xlabel('Monitoring Window',fontsize=15)
plt.ylabel('Average Bandwith (M/seg)',fontsize=15)
plt.savefig('bandwithperobswin.pdf')
```
