```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np

### Statistics each 109 seconds, corresponding to 1 observation window ###

### Statistics per observation window ### SW51
stats51entry = pd.read_csv("stats51entry.txt",delimiter='|',usecols=[3])
stats51 = pd.read_csv("stats51.txt",delimiter='|',usecols=[3])

plt.figure(figsize=(10, 5))
plt.suptitle('Victim`s Switch Bandwith Utilization', fontsize=14)
plt.plot(stats51/100000,color='#C2185B',lw=1,label='Throughput')

plt.errorbar((3000,),(3.46,), xerr=750, ecolor='black', linestyle='-.',  mew=1,capsize=4, label=None, lw=1)
plt.text(3000, 3.25,'Attack', ha='center', va='bottom', color='black', fontsize=11)

plt.xlabel('Monitoring Window',fontsize=13)
plt.ylabel('Average Bandwith (M/seg)',fontsize=13)
plt.legend(loc='center left')
plt.savefig('bandwith51.pdf')
```
