```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
print ('####  SWITCH 11 WHEN ATTACK IS PRESENT ####')
       
s1sc1 = np.loadtxt('source1scen1.txt')
srcs1sc1 = s1sc1[:,1]
dsts1sc1 = s1sc1[:,2]
#srcs1sc1a = s1sc1[500:749,1]
#dsts1sc1a = s1sc1[500:749,2]

longs1sc1 = len(srcs1sc1)
indexs1sc1 = range(1,longs1sc1+1)
#longs1sc1a = len(srcs1sc1a)
#indexs1sc1a = range(1,longs1sc1a+1)

plt.figure(figsize=(16, 10))
plt.scatter(indexs1sc1,srcs1sc1 / 16, color='gray', lw=0.00001)
plt.scatter(indexs1sc1,dsts1sc1 / 16, color='black', lw=0.01)
#plt.scatter(indexs1sc1a,srcs1sc1a / 16, color='red', lw=0.00001)
#plt.scatter(indexs1sc1a,dsts1sc1a / 16, color='red', lw=0.01)
plt.legend(['SourceEntropy','Destination Entropy'])
plt.xlabel('Observation Window')
plt.ylabel('Entropy Value')
plt.show()
```
